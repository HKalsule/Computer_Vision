```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# 1. Load your custom audio file
file_path = 'SoundAudio.wav'
y, sr = librosa.load(file_path)

# 2. Basic Info
duration = librosa.get_duration(y=y, sr=sr)
print(f"Sample Rate: {sr} Hz | Duration: {duration:.2f} sec")

# 3. Plot Waveform
plt.figure(figsize=(12, 4))
librosa.display.waveshow(y, sr=sr)
plt.title('Speech Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

# 4. Plot Spectrogram
X = librosa.stft(y)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(12, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz', cmap='magma')
plt.colorbar(label='dB')
plt.title('Spectrogram (Frequency vs Time)')
plt.tight_layout()
plt.show()

# 5. Short-Time Energy Plot
frame_length = 2048
hop_length = 512
energy = np.array([
    np.sum(np.abs(y[i:i+frame_length]**2))
    for i in range(0, len(y), hop_length)
])

plt.figure(figsize=(10, 4))
frames = range(len(energy))
t = librosa.frames_to_time(frames, sr=sr, hop_length=hop_length)
plt.plot(t, energy, color='red')
plt.title('Short-Time Energy')
plt.xlabel('Time (s)')
plt.ylabel('Energy')
plt.grid(True)
plt.tight_layout()
plt.show()

# 6. Pitch (F0) Estimation
f0, voiced_flag, voiced_probs = librosa.pyin(y, fmin=50, fmax=500, sr=sr)
plt.figure(figsize=(10, 4))
plt.plot(librosa.times_like(f0), f0, label='Pitch (Hz)', color='green')
plt.title('Estimated Pitch (Fundamental Frequency)')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
